## Modeling

The following models are my first simple models

```python
# Import libraries
import time, scipy
import matplotlib.pyplot as plt
from sklearn.metrics import recall_score
import pickle

from keras import models
from keras import layers
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, MaxPooling2D, Activation, Conv2D, Dropout
from tensorflow.keras.optimizers import SGD

# .py file
from Helpers import helpers as hp
```

```python
# Load Data
X = pickle.load(open('Pickles/X_val.pickle', 'rb'))
y = pickle.load(open('Pickles/y_val.pickle', 'rb'))
```

```python
# First simple model
model = models.Sequential() # Model steps are performed one after another rather than at the same time
model.add(Flatten()) # Transform features into a single column
model.add(layers.Dense(20, activation='relu', input_shape=X.shape[1:])) # First hidden layer
model.add(layers.Dense(15, activation='relu')) # Second hidden layer
model.add(layers.Dense(10, activation='relu')) # Third hidden layer
model.add(layers.Dense(3, activation='softmax')) # Output layer, 1 node for each class

opt = SGD(learning_rate=0.01)
model.compile(loss = 'sparse_categorical_crossentropy',
              optimizer = opt,
              metrics=['accuracy']
)

model_score = model.fit(X,
                        y,
                        epochs=50,
                        validation_split=0.2)
```

```python
# Model performance per epoch
hp.plot_model_train_val_accuracy_loss(model_score)
```

```python
data_augmentation = Sequential([ # Model steps are performed one after another rather than at the same time
layers.experimental.preprocessing.RandomFlip('horizontal_and_vertical'),
layers.experimental.preprocessing.RandomRotation(0.2)
])
```

```python
# Model with augmented data to account for class imbalance
model3 = Sequential([ # Model steps are performed one after another rather than at the same time
  data_augmentation,
  layers.Conv2D(filters=48, kernel_size=3, activation='relu', input_shape=X.shape[1:]),
  layers.MaxPooling2D(pool_size=2, strides=2)
])
model3.add(layers.Flatten())
model3.add(layers.Dense(128, activation='relu'))
model3.add(layers.Dense(64, activation='relu'))
model3.add(layers.Dense(3, activation='softmax'))

# finally compile and train the cnn
model3.compile(loss = 'sparse_categorical_crossentropy',
               optimizer = opt,
               metrics=['accuracy'])

model3_score = model3.fit(X,
                          y,
                          epochs=5, # Few epochs due to computational time
                          validation_split=0.2)
```

```python
# View model performance per epoch
hp.plot_model_train_val_accuracy_loss(model3_score)
```

```python
# Model with additional hidden layers
# Initialize model
model4 = Sequential([ # Model steps are performed one after another rather than at the same time
  data_augmentation,
  layers.Conv2D(filters=48, kernel_size=3, activation='relu', input_shape=X.shape[1:]),
  layers.MaxPooling2D(pool_size=2, strides=2)
])
model4.add(layers.Flatten())
# Input layer with dropout
model4.add(layers.Dense(15, activation='relu'))
model4.add(layers.Dense(10, activation='relu')) # First hidden layer
model4.add(layers.Dropout(0.3))
model4.add(layers.Dense(10, activation='relu')) # Second hidden layer
model4.add(layers.Dense(10, activation='relu')) # Third hidden layer
model4.add(layers.Dropout(0.3))
model4.add(layers.Dense(3, activation='softmax')) # Output layer

model4.compile(loss = 'sparse_categorical_crossentropy',
               optimizer = opt,
               metrics=['accuracy'])

model4_score = model4.fit(X,
                          y,
                          epochs=10,
                          validation_split=0.2)
```

```python
# Model performance per epoch
hp.plot_model_train_val_accuracy_loss(model4_score)
```
