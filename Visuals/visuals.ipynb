```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
count = [39.5, 25.5, 11.0, 9.6, 9.3, 8.6, 2.3, 1.7]
reason = ['Nudity and sexual activity', 'Violence', 'Dangerous organizations', 'Hate speech', 'Drugs or firearms', 'Child Exploitation', 'Bullying', 'Suicide and self harm']



# plot churn for each additional phone call
fig_dims = (16,8)
fig, ax = plt.subplots(figsize=fig_dims)
graph = plt.bar(x=reason, 
        height=count,
        width = 0.6, 
        tick_label=reason,
        color=(0.2, 0.4, 0.6, 0.6))
#plt.xticks(rotation=65)

for index,data in enumerate(count):
    plt.text(x=index , 
             y =data+.5 , 
             s=f"{data}" , 
             fontdict=dict(fontsize=11)) # plot percentages over bars

# assign x and y labels
plt.title('Posts Deleted on Facebook In First Quarter 2020', size=16)
plt.ylabel('Posts Deleted In Millions', size=14)
plt.xlabel('Reason For Deletion', size=14)
# increase size of tick labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
ax.yaxis.grid(False) # Hide the horizontal gridlines
ax.xaxis.grid(False) # Hide the vertical gridlines
    
plt.show()
```
