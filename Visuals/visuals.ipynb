```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
count = [39.5, 25.5, 11, 9.6, 9.3, 8.6, 2.3, 1.7]
reason = ['Nudity and sexual activity', 'Violence', 'Dangerous organizations', 'Hate speech', 'Drugs and firearms', 'Exploitation of children', 'Bullying', 'Suicide and self harm']



# plot churn for each additional phone call
sns.set()
plt.figure(figsize=(12,8))
plt.bar(x=reason, 
        height=count,
        width = 0.6, 
        tick_label=reason,
        color='darkred')
plt.xticks(rotation=65)

for index,data in enumerate(count):
    plt.text(x=index , 
             y =data+1 , 
             s=f"{data}" , 
             fontdict=dict(fontsize=11)) # plot percentages over bars

# assign x and y labels
plt.title('Posts Deleted on Facebook In First Quarter 2020', size=16)
plt.ylabel('Posts Deleted In Millions', size=14)
plt.xlabel('Reason For Deletion', size=14)
# increase size of tick labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
    
plt.show()
```
